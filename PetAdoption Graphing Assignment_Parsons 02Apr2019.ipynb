Can you compare two related quantities?  The last two plots in this section compard the AdoptionSpeed
with the Dewormed and Breed1 Attribute and can be used to answer this question.  

The target variable for this study is Adoption Speed and is an ordinal categorical variable. 
A count plot was initated to see the relationships between the rate at which animals were 
adopted from the shelter.  The majority of the animnals were adopted after the first seven 
days at a consistent rate as shown in the histogram count plot below.  The bars can be misleading 
because the number of days between each bar is not uniform or the same.  The 0 category is for 
animals adopted the same day they came in the shelter.  Category 1 is for animals that have been 
in the shelter for only seven days and then were adopted.  Category 2 is for animals 
that were in the shelter for only 30 days before being adopted.  Animals that were in the shelter 
for 31 to 90 days were a category 3 and any animal that was in the shelter for more than 100 days 
was category 4.  Category two has four times the number of days compared to one and three has more
than eigth times the number of days compared to category 1.  I will make the initial comparisons 
using an ordinal target variable, but most likely will bin this into a binary category.  The first 
three levels will be in the first bin and the other two will be placed in the second bin.  

Two Histograms below are count plots are using the AdoptionSpeed as the x-axis and 
Dewormed and Color1 as the hue to group the columns.  These two histograms show that animals 
that were dewormed and had Color1 had a higher chance of being adopted for all five AdoptionSpeed 
categories.  These two factors were consistently higher as AdoptionSpeed increased over time.  The 
animals that had color 1 or Black was the best color option for animals to be adopted 
and color 2 was brown.  Animals that had the darker pigmented fur color had a better chance for 
being adopted in this training data set.  The remaining animals with a coat color of 3 to 7 were not
adopted as frequently, but this also only made up about 5,000 of the sample points in this study.  

The last histogram shows the type of animal (1 is for dog and 2 is for cat) being adopted is not 
influenced by the health of the animal in the shelter.  Looking at the graph, only around 500 
out of 14,993 animals were were listed as a 2 or 3, which was a minor to serious injury.  
The majority of the animals were healthy and this had no affect on adoption.

```python
# Correlation Matrix Plot

names=['Adop','Brd1', 'Age', 'Type', 'FurL', 'Ster', 
                                'Quan', 'Vacc', 'Gen']
correlations = data_I.corr()
# plot correlation matrix
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,9,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names)
ax.set_yticklabels(names)
plt.show()
```
